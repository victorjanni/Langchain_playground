```python
from langchain_openai import OpenAIEmbeddings
embeddings=OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text="this is a tutorial of openai embeddings"
query_result=embeddings.embed_query(text)
query_result
```

```python
embeddings_1012=OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1012)
```

```python
embeddings_1012
```

```python
text="this is a 1012 demensions embeddings"
query_result=embeddings_1012.embed_query(text)
query_result
```

```python
#text loader
from langchain_community.document_loaders import TextLoader
loader=TextLoader('speech.txt')
docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_documents=text_splitter.split_documents(docs)
final_documents
```

```python
len(final_documents)
```

```python
## vector embeddings and vector store
from langchain_community.vectorstores import Chroma

db=Chroma.from_documents(final_documents,embeddings_1012)
db
```

```python
## retrieve the results from query vectorstore db
query="To such a task we can dedicate our lives and our fortune"
retrieved_results=db.similarity_search(query)
print(retrieved_results)
```
