#### Ollama
Ollama supports embedding models, making it possible to build retrieval augmented generation (RAG) applications that combine text prompts with existing documents or other data.

```python
from langchain_ollama import OllamaEmbeddings
```

```python
embeddings=(
    OllamaEmbeddings(model="nomic-embed-text") #by default use llama2
)
```

```python
embeddings
```

```python
r1=embeddings.embed_documents(
    [
        "deep learning is the subset of machine learning",
        "machine learning is the subset of artificial intelligence",
    ]
)
```

```python
len(r1[0])
```

```python
r1
```

```python
embeddings.embed_query("what is the subset of machine learning")
```

```python
### Other Embedding Models
### https://ollama.com/blog/embedding-models
embeddings = OllamaEmbeddings(model="mxbai-embed-large")
text = "This is a test document."
query_result = embeddings.embed_query(text)
query_result
```
